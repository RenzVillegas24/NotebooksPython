```python
import os
class POSmachine:
    
    def __init__(self):
        """
        Initialize the fields.
        
        Make sure that the last line of this method is the following:
        self.start()
        """
        # Initialize your fields here
        self.list = []
        self.start()
        
    
    def start(self):
        while True:
            self.printWelcomeMessage()
            self.printAllOptions()
            inp = input()

            if inp == '1':
                print()
                print('Input a file path: \n', end="")       
                flPath = input()
                print(flPath) 
                lst = self.setPriceList(flPath)
                
                if lst:
                    self.list = lst
                    print("\nSuccess!\nFile (%s) successfully loaded!" % (flPath))
                else:
                    print("\nError!\nFile (%s) not found in the given directory!" % (flPath))
                print()
            elif inp == '2':
                self.printPriceList()
            elif inp == '3':
                self.makeNewTransaction()
            elif inp == '4':
                break
            else:
                continue

                
    def printWelcomeMessage(self):
        print()
        print("WELCOME to the POSmachine!")
            
    def printAllOptions(self):
        print("What do you want to do today? (input the number)")
        print("1 - Set the price list")
        print("2 - Show the price list")
        print("3 - Make a new transaction")
        print("4 - Close the POS machine")
        print()

    
    def setPriceList(self,file_path):
        if not os.path.isfile(file_path): 
            return False

        file_ext = os.path.splitext(file_path)[1]
        file_del = ''
        if file_ext == ".txt": 
            file_del = '\t'
        elif file_ext == ".csv": 
            file_del = ','
        else: 
            return False


        file = open(file_path, "r")
        file_r = file.read().splitlines()
        file_raw = []

        for string in file_r:
            tmp = ''
            hasQuote = False
            for character in string:
                if not hasQuote and character == file_del:
                    character = '\n'
                elif character == '"':
                    hasQuote = not hasQuote
                tmp += character
            file_raw.append(tmp.split('\n'))

        return file_raw

        
    
    def printPriceList(self):
        mxSz = 0    
        for col in self.list:
            if len(col[1]) > mxSz:
                mxSz = len(col[1])
        
        frmt = "{0:4}{1:%i}{2:7}" % (mxSz + 4)
        print('')
        print("Price List:")
        for col in self.list:
            print(frmt.format(col[0],col[1],col[2]))
        print('')

        
    
    def addProductToTransaction(self, product, quantity):
        """
        Calculate the price for the given product and quantity
        and add this to the current invoice
        
        Remove/delete the "pass" keyword below.
        """
        # Add your code here
        pass
    
    def makeNewTransaction(self):
        """
        Create a new transaction. Print the price list once.
        Have the user to add products (given a quantity >= 1).
        When the user inputs a blank, print the invoice.
        
        Remove/delete the "pass" keyword below.
        """
        # Add your code here
        pass
    
    def createSalesInvoice(self):
        """
        Print (into the console) the sales invoice, detailing
        the price and quanity of the products purchased, and
        the total amount.
        
        Save the same print output as a text file.
        
        Remove/delete the "pass" keyword  below.
        """
        # Add your code here
        pass
```
